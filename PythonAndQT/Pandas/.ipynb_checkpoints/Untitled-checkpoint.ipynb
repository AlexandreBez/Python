```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Empresa':['A', 'B', 'C', 'D'],
        'Setor':['Bebidas', 'Carnes', 'Paes', 'Frutas'],
        'Vendedores':['Aline', 'Joao', 'Flavia', 'Paulo'],
        'Venda':[100,200,300,400]},
        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'Empresa':['E', 'F', 'G', 'H'],
        'Setor':['Legumes', 'Doces', 'Vinhos', 'Massas'],
        'Vendedores':['Aline', 'Joao', 'Flavia', 'Paulo'],
        'Venda':[100,200,300,400]},
        index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'Empresa':['I', 'J', 'K', 'L'],
        'Setor':['Congelados', 'Importados', 'Enlatados', 'Almoco'],
        'Vendedores':['Aline', 'Joao', 'Flavia', 'Paulo'],
        'Venda':[100,200,300,400]},
        index=[8, 9, 10, 11])
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], axis = 1)
```

```python
esquerda = pd.DataFrame({'Empresa':['I', 'J', 'K', 'L'],
        'Setor':['Congelados', 'Importados', 'Enlatados', 'Almoco'],
        'Vendedores':['Aline', 'Joao', 'Flavia', 'Paulo'],
        'Venda':[100,200,300,400]},
        index=[8, 9, 10, 11])
```

```python
direita = pd.DataFrame({'Empresa':['E', 'F', 'G', 'H'],
        'Setor':['Legumes', 'Doces', 'Vinhos', 'Massas'],
        'Vendedores':['Aline', 'Joao', 'Flavia', 'Paulo'],
        'Venda':[100,200,300,400]},
        index=[4, 5, 6, 7])
```

```python
pd.merge(esquerda,  direita, how = 'inner', on = 'Empresa')
```
